## 1. Loading the dataset

File Descriptions
- train.zip: contains 2 csvs and 1 folder containing image data
- train.csv – [‘image_names’, ‘class’] contains the image name and correct class for 12196 train images
- image – contains images for both train and test sets
- test.csv: [‘image_names’] contains just the image names for 5227 images in the test set
- sample_submission.csv: [‘image_names’,’class’] contains the exact format for a valid submission

```python
# import necessary libraries and functions
import numpy as np
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
%matplotlib inline

from keras.layers import Dense, InputLayer
from keras.models import Sequential

from sklearn.model_selection import train_test_split
```

```python
# reading the csv file
data = pd.read_csv('../../data/train_gender.csv')
```

```python
# create random number generator
seed = 42
rng = np.random.RandomState(seed)
```

```python
# looking at first five rows of the data
data.head()
```

```python
# distribution of classes in the dataset
# Males are identified with a value of 1
data['class'].value_counts()
```

```python
# load images and store it in numpy array

# empty list to store the images
X = []
# iterating over each image
for img_name in data.image_names:
    # loading the image using its name
    img = plt.imread('../../data/images_gender/' + img_name)
    # saving each image in the list
    X.append(img)
```

```python
# converting the list of images into array
X=np.array(X)
```

```python
# shape of the image
X.shape
```

```python
# storing the target variable in separate variable
y = data['class'].values
```

```python
# visualizing images
fig,ax = plt.subplots(nrows=1,ncols=5,figsize=(20,20))
for i in range(5):
    ax[i].imshow(X[i*400])
    ax[i].axis('off')
```

## 2. Pre-processing the data

```python
# converting 3 dimensional image to 1 dimensional image
X = X.reshape(X.shape[0], 224*224*3)
X.shape
```

```python
# minimum and maximum pixel values of images
X.min(), X.max()
```

```python
# normalizing the pixel values
# This reduces training time
X = X / X.max()
```

```python
# minimum and maximum pixel values of images after normalizing
X.min(), X.max()
```

## 3. Creating training and validation set

```python
# creating a training and validation set
X_train, X_valid, y_train, y_valid=train_test_split(X,y,test_size=0.20, random_state=seed)
```

```python
# shape of training and validation set
(X_train.shape, y_train.shape), (X_valid.shape, y_valid.shape)
```

X has 1646 images and the test set has 706.

```python
# defining the architecture of the model
model=Sequential()
model.add(InputLayer(input_shape=(X_train.shape[1],)))
model.add(Dense(1024, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
#Accuracy on validation set: 
```

```python
# summary of the model
model.summary()
```

## 5. Compiling the model

```python
# compiling the model
#model.compile(loss='binary_crossentropy', optimizer="sgd", metrics=['accuracy'])
# #Accuracy on validation set: 0.7229508196721312 % original
model.compile(loss='binary_crossentropy', optimizer="adam", metrics=['accuracy'])
```

## 6. Training the model

```python
# training the model
model_history = model.fit(X_train, y_train, epochs=100, batch_size=128,validation_data=(X_valid,y_valid)) # was 50 before 2048 layer
```

## 7. Evaluating model performance

```python
# getting predictions in the form of class as well as probabilities
#predictions = model.predict_classes(X_valid)[:, 0] #Old code that needed to be changed.  Run to see warning.
predictions = (model.predict(X_valid) > 0.5).astype("int32")

prediction_probabilities = model.predict(X_valid)[:, 0]
```

```python
# accuracy on validation set
from sklearn.metrics import accuracy_score
print('Accuracy on validation set:', accuracy_score(y_valid, predictions), '%')
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
# summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
