```python
# importing the required libraries
%matplotlib inline
import numpy as np  
import pandas as pd 
from tqdm import tqdm
import matplotlib.pyplot as plt 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# importing libraries for defining the architecture of model
import torch 
import torch.nn as nn
from torch.optim import Adam
from torch.autograd import Variable
from torch.nn import Linear, ReLU, BCELoss, Sequential, Sigmoid, Dropout, BatchNorm1d, Conv2d, MaxPool2d, BatchNorm2d
```

```python
#reading a csv file
data = pd.read_csv('../data/emergency_vehicles/emergency_classification.csv')
```

```python
# creating a random seed
seed = 42
```

```python
#first 5 rows
data.head()
```

```python
# load images and store it in numpy array

# empty list to store the images
X = []
# iterating over each image
for img_name in data.image_names:
    # loading the image using its name
    img = plt.imread('../data/emergency_vehicles/images_vehicles/' + img_name)
    # normalizing the pixel values
    img = img / 255
    # saving each image in the list
    X.append(img)
    
# converting the list of images into array
X=np.array(X)

# storing the target variable in separate variable
y = data.emergency_or_not.values

#Glance at the images
fig, axes = plt.subplots(ncols=5,figsize=(20,10))
for i in range(5):
  axes[i].imshow(X[i])
```

## 2. Pre-processing the Data

```python
# shape of images
X.shape
```

```python
#reshaping array as pytorch acceps channels as the first dimension of an image
X = np.moveaxis(X, -1, 1)  
X.shape
```

## 3. Creating Training and Validation set

```python
# creating a training and validation set
X_train, X_valid, y_train, y_valid=train_test_split(X,y,test_size=0.3, random_state=seed, stratify=y)
```

```python
# shape of training and validation set
(X_train.shape, y_train.shape), (X_valid.shape, y_valid.shape)
```

```python
# converting training and validation set to PyTorch tensor

X_train = torch.FloatTensor(X_train)
y_train = torch.FloatTensor(y_train)

X_valid = torch.FloatTensor(X_valid)
y_valid = torch.FloatTensor(y_valid)
```

## 4. Defining the model architecture

```python
from IPython.display import Image
Image(filename='calc.png')
```

[Calculating Parameters of Convolutional and Fully Connected Layers with Keras](https://dingyan89.medium.com/calculating-parameters-of-convolutional-and-fully-connected-layers-with-keras-186590df36c6)

Below is an equation for calculating the output of convolutional layers as follows:

Narrative example:
The input shape is (32,32,3), kernel size of first Conv Layer is (5,5), with no padding, the stride is 1, so the output size is (32–5)+1=28. And the number of filters is 8. So the output shape of the first Conv layer is (28,28,8). Followed by a max-pooling layer, the method of calculating pooling layer is as same as the Conv layer. The kernel size of max-pooling layer is (2,2) and stride is 2, so output size is (28–2)/2 +1 = 14. After pooling, the output shape is (14,14,8). 

```python
# Define CNN architecture

# Input 
# Conv2d(64, kernel_size=(3, 3), stride=(1, 1), 'relu')
# MaxPool2d((2, 2), stride=2)
# Conv2d(128, kernel_size=(3, 3), stride=(1, 1), 'relu')
# MaxPool2d((2, 2), stride=2)
# Linear(64, 'relu')
# Linear(1, 'sigmoid')

# Recall in PyTorch to create a layer we must define the properties of the previous layer manually
# The propoerties are othing but the features of the previous layer
    # The number of channels in the input layer, 
    # number of filters in the convulational layer, 
    # or number of nearons in the dense layers

class Net(nn.Module):   
  def __init__(self):
      super(Net, self).__init__()

      #define the convolution block
      self.conv_layers = Sequential(
         
         #[bs,3,224,224] 
         Conv2d(in_channels=3, out_channels=64, kernel_size=3),
          # in_channels - the number of channels
          # out_channels - the number of fitlers
          # kernel_size - size of the filters
         
         #[bs,64,222,222]
         ReLU(),
         # kernel_size is just the pooling size
          
         #[bs,64,222,222]
         MaxPool2d(kernel_size=2, stride=2),
         
         #[bs,64,111,111]
         Conv2d(64, 128, 3),

         #[bs,128,109,109]
         ReLU(),

         #[bs,128,109,109]
         MaxPool2d(2, 2)    # kernel_size and stride      
         
         )
      
        # input -> [bs,3,224,224] bs stands for batch size
        
        # Conv2d(64, kernel_size=(3, 3), stride=(1, 1), 'relu') -> [bs,64,222,222] 
            # 222 are the height and width [(224-3)+1]
            # https://keras.io/api/layers/pooling_layers/max_pooling2d/
      
        # MaxPool2d((2, 2), stride=2) -> [bs,64,111,111]
            # output_shape = math.floor((input_shape - pool_size) / strides) + 1
            # output_shape = ((222-2)/2)+1 = 110 + 1 = 111

        # Conv2d(128, kernel_size=(3, 3), stride=(1, 1), 'relu') -> [bs,128,109,109]
            # (111 - 3) + 1 = 108 + 1 = 109

        # MaxPool2d((2, 2), stride=2) -> [bs,128,54,54]
            # output_shape = math.floor((input_shape - pool_size) / strides) + 1 
            # ((109 - 2)/2) + 1 = 53 + 1 = 54

        # Flatten -> [bs, 128*54*54]


      #define the dense block   
      self.linear_layers=Sequential(
           
           #[bs,128*54*54]
           Linear(128 * 54 * 54, 64), # 64 is the number of neurons

           #[bs,128*54*54]
           ReLU(),

           #[bs,64]
           Linear(64, 1),
           
           #[bs,1]
           Sigmoid()
       )
  
  #Defining the forward pass    
  def forward(self, x):
    
    #convolution block
    x = self.conv_layers(x)
    
    #flattening the output
    x = x.view(x.size(0),-1)
    
    #dense block
    x = self.linear_layers(x)
    
    return x
```

## 5. Compiling the model (defining loss function, optimizer)

```python
# for same default initialization
torch.manual_seed(32)

# defining the model
model = Net()

#define optimizer and loss
optimizer = Adam(model.parameters(), lr=0.001)
criterion = nn.BCELoss()

# checking if GPU is available
if torch.cuda.is_available():
    model = model.cuda()
    criterion = criterion.cuda()
    
print(model)
```

## 6. Training the model

```python
# batch size of the model
batch_size = 64

for epoch in tqdm(range(50)):
    # keep track of training and validation loss
    train_loss = 0.0
        
    permutation = torch.randperm(X_train.size()[0])

    training_loss = []
    for i in range(0,X_train.size()[0], batch_size):
        model.train()
        indices = permutation[i:i+batch_size]
        batch_x, batch_y = X_train[indices], y_train[indices]
        
        if torch.cuda.is_available():
            batch_x, batch_y = batch_x.cuda(), batch_y.cuda()
        
        optimizer.zero_grad()
        # batch computation
        outputs = model(batch_x)
        loss = criterion(outputs.squeeze(),batch_y)

        training_loss.append(loss.item())
        loss.backward()
        optimizer.step()
        
    training_loss = np.average(training_loss)
    if epoch%10 == 0:
        # printing the loss
        print('epoch: \t', epoch, '\t training loss: \t', training_loss)
```

## 7. Evaluating model performance

```python
# prediction for training set
train_prediction = []
train_target = []
permutation = torch.randperm(X_train.size()[0])
for i in tqdm(range(0,X_train.size()[0], batch_size)):
    model.eval()
    indices = permutation[i:i+batch_size]
    batch_x, batch_y = X_train[indices], y_train[indices]

    if torch.cuda.is_available():
        batch_x, batch_y = batch_x.cuda(), batch_y.cuda()

#     with torch.no_grad():
#         output = model(batch_x.cuda()).data.cpu()

    with torch.no_grad(): # prevent creating gradients during the eval phase
        output = model(batch_x).data.cpu()
    
    # converting the predicted probabilities into labels
    train_predicted_label = []
    for i in output:
        # defining the threshold value as 0.5
        if i>=0.5:
            train_predicted_label.append(1)
        else:
            train_predicted_label.append(0)
    train_prediction.append(train_predicted_label)
    train_target.append(batch_y)
```

```python
# training accuracy
training_accuracy = []
for i in range(len(train_prediction)):
    training_accuracy.append(accuracy_score(train_target[i].data.cpu(),train_prediction[i]))
    
print('training accuracy: \t', np.average(training_accuracy))
```

```python
# prediction for validation set
valid_prediction = []
valid_target = []
permutation = torch.randperm(X_valid.size()[0])
for i in tqdm(range(0,X_valid.size()[0], batch_size)):
    model.eval()
    indices = permutation[i:i+batch_size]
    batch_x, batch_y = X_valid[indices], y_valid[indices]

    if torch.cuda.is_available():
        batch_x, batch_y = batch_x.cuda(), batch_y.cuda()

#     with torch.no_grad():
#         output = model(batch_x.cuda()).data.cpu()
        
    with torch.no_grad(): # prevent creating gradients during the eval phase
        output = model(batch_x).data.cpu()
    
    # converting the predicted probabilities into labels
    valid_predicted_label = []
    for i in output:
        # defining the threshold value as 0.5
        if i>=0.5:
            valid_predicted_label.append(1)
        else:
            valid_predicted_label.append(0)
    valid_prediction.append(valid_predicted_label)
    valid_target.append(batch_y)
```

```python
# validation accuracy
valid_accuracy = []
for i in range(len(valid_prediction)):
    valid_accuracy.append(accuracy_score(valid_target[i].data.cpu(),valid_prediction[i]))
    
print('validation accuracy: \t', np.average(valid_accuracy))
```

## 8. Hyperparameter Tuning

```python
# defining the architecture of the model

# 4 convolutional blocks.  Each followed by a batch layer, ReLU layer and a pooling layer.

# 1. Input()
# 2. Conv2d(64, kernel_size=(3, 3), stride=(1, 1))
# 3. BatchNorm2d()
# 4. ReLU()
# 5. MaxPool2d(kernel_size=2, stride=2)
# 6. Conv2d(128, kernel_size=(3, 3), stride=(1, 1))
# 7. BatchNorm2d()
# 8. ReLU()
# 9. MaxPool2d(kernel_size=2, stride=2)
# 10. Conv2d(128, kernel_size=(5, 5), stride=(1, 1))
# 11. BatchNorm2d()
# 12. ReLU()
# 13. MaxPool2d(kernel_size=2, stride=2)
# 14. Conv2d(256, kernel_size=(5, 5), stride=(1, 1))
# 15. BatchNorm2d()
# 16. ReLU()
# 17. MaxPool2d(kernel_size=2, stride=2)
# 18. Flatten()
# 19. Linear(64, 'relu')
# 20. Dropout(p=0.6)
# 21. Linear(out_features=1, 'sigmoid')

class Net(nn.Module):   
  def __init__(self):
      super(Net, self).__init__()

      #define the convolution block
      self.conv_layers = Sequential(
                  #[bs,3,224,224]: starting data input
         
        # 2. Conv2d(64, kernel_size=(3, 3), stride=(1, 1))
                  Conv2d(in_channels=3, out_channels=64, kernel_size=3),
                  #[bs,64,222,222]: 64 is model design; 222 = ((224-3 + 2*0)/1)+1 
         
        # 3. BatchNorm2d()
                  BatchNorm2d(num_features=64), #64 comes from the out_channels in the layer above
                  #[bs,64,222,222]: no change
         
        # 4. ReLU()
                  ReLU(),
                  #[bs,64,222,222]: no change
                 
        # 5. MaxPool2d(kernel_size=2, stride=2)
                  MaxPool2d(kernel_size=2, stride=2),
                  # [64, 111, 111]:((222-2)/2)+1=111

        # 6. Conv2d(128, kernel_size=(3, 3), stride=(1, 1))
                  Conv2d(in_channels=64, out_channels=128, kernel_size=3), # model design
                  # [128, 109, 109]: ((111-3)/1)+1=109

        # 7. BatchNorm2d()
                  BatchNorm2d(num_features=128),
         
        # 8. ReLU()
                  ReLU(),
                  #[bs,128,109,109]: no change
         
         # 9. MaxPool2d(kernel_size=2, stride=2)
                  MaxPool2d(2, 2),
                  #[bs,128,54,54]: ((109-2)/2)+1 = floor(54.5) = 54
          
         # 10. Conv2d(128, kernel_size=(5, 5), stride=(1, 1))
                  Conv2d(in_channels=128, out_channels=128, kernel_size=5), # model design
                  #[bs,128,54,54]: 
         
         # 11. BatchNorm2d()
                  BatchNorm2d(num_features=128),
                  #[bs,128,54,54]: no change
         
         # 12. ReLU()
                  ReLU(),
                  #[bs,128,54,54]: no change
         
         # 13. MaxPool2d(kernel_size=2, stride=2)
                  MaxPool2d(2, 2), 
                  #[bs,128,25,25]: ((54-2)/2)+1=  ?????????????????????????????????????

        # 14. Conv2d(256, kernel_size=(5, 5), stride=(1, 1))
                 Conv2d(in_channels=128, out_channels=256, kernel_size=5),
                 # output_size = strides * (input_size-1) + kernel_size - 2*padding
                 # 2*(54-1) + 2 - 2*0 = 106
          
        # 15. BatchNorm2d()
                  BatchNorm2d(num_features=256),
                  #[bs,256,21,21]: 25-5+1 = 21
        
          # 16. ReLU()
                  ReLU(), 

          # 17. MaxPool2d(kernel_size=2, stride=2)
                  MaxPool2d(2, 2)
                  # ((21-2)/2 +1 = 10
         )
        
        # 18. Flatten()
      
      self.linear_layers=Sequential(
           
        # 18. Flatten()
                  Linear(10 * 10 * 256, 64),
                  # 64 neurons was specified
                
           
        # 19. Linear(64, 'relu')
                  ReLU(),
                  #[bs,64]
           
          # 20. Dropout(p=0.6)
                  Dropout(0.6),
           
          # 21. Linear(out_features=1, 'sigmoid')
                  Linear(64, 1),
                  Sigmoid()
       )
  
  # Defining the forward pass    
  def forward(self, x):
    
    #convolution block
    x = self.conv_layers(x)
    
    #flattening
    x = x.view(x.size(0),-1)
    
    #dense block
    x = self.linear_layers(x)
    
    return x
```

```python
# for same default initialization
torch.manual_seed(32)

# defining the model
model = Net()

#define optimizer and loss
optimizer = Adam(model.parameters(), lr=0.001)
criterion = nn.BCELoss()

# checking if GPU is available
if torch.cuda.is_available():
    model = model.cuda()
    criterion = criterion.cuda()
    
print(model)
```

```python
# batch size of the model
batch_size = 128

for epoch in tqdm(range(50)):
    # keep track of training 
    train_loss = 0.0
        
    permutation = torch.randperm(X_train.size()[0])

    training_loss = []
    for i in range(0,X_train.size()[0], batch_size):
        model.train()
        indices = permutation[i:i+batch_size]
        batch_x, batch_y = X_train[indices], y_train[indices]
        
        if torch.cuda.is_available():
            batch_x, batch_y = batch_x.cuda(), batch_y.cuda()
        
        optimizer.zero_grad()
        # batch computation
        outputs = model(batch_x)
        loss = criterion(outputs.squeeze(),batch_y)

        training_loss.append(loss.item())
        loss.backward()
        optimizer.step()
        
    training_loss = np.average(training_loss)
    if epoch%10 == 0:
        # printing the loss
        print('epoch: \t', epoch, '\t training loss: \t', training_loss)
```

```python
# prediction for training set
train_prediction = []
train_target = []
permutation = torch.randperm(X_train.size()[0])
for i in tqdm(range(0,X_train.size()[0], batch_size)):
    model.eval()
    indices = permutation[i:i+batch_size]
    batch_x, batch_y = X_train[indices], y_train[indices]

    if torch.cuda.is_available():
        batch_x, batch_y = batch_x.cuda(), batch_y.cuda()

#     with torch.no_grad():
#         output = model(batch_x.cuda()).data.cpu()

    with torch.no_grad(): # prevent creating gradients during the eval phase
        output = model(batch_x).data.cpu()
    
    # converting the predicted probabilities into labels
    train_predicted_label = []
    for i in output:
        # defining the threshold value as 0.5
        if i>=0.5:
            train_predicted_label.append(1)
        else:
            train_predicted_label.append(0)
    train_prediction.append(train_predicted_label)
    train_target.append(batch_y)
```

```python
# training accuracy
training_accuracy = []
for i in range(len(train_prediction)):
    training_accuracy.append(accuracy_score(train_target[i].data.cpu(),train_prediction[i]))
    
print('training accuracy: \t', np.average(training_accuracy))
```

```python
# prediction for validation set
valid_prediction = []
valid_target = []
permutation = torch.randperm(X_valid.size()[0])
for i in tqdm(range(0,X_valid.size()[0], batch_size)):
    model.eval()
    indices = permutation[i:i+batch_size]
    batch_x, batch_y = X_valid[indices], y_valid[indices]

    if torch.cuda.is_available():
        batch_x, batch_y = batch_x.cuda(), batch_y.cuda()

#     with torch.no_grad():
#         output = model(batch_x.cuda()).data.cpu()
        
        with torch.no_grad(): # prevent creating gradients during the eval phase
            output = model(batch_x).data.cpu()
    
    # converting the predicted probabilities into labels
    valid_predicted_label = []
    for i in output:
        # defining the threshold value as 0.5
        if i>=0.5:
            valid_predicted_label.append(1)
        else:
            valid_predicted_label.append(0)
    valid_prediction.append(valid_predicted_label)
    valid_target.append(batch_y)
```

```python
# validation accuracy
valid_accuracy = []
for i in range(len(valid_prediction)):
    valid_accuracy.append(accuracy_score(valid_target[i].data.cpu(),valid_prediction[i]))
    
print('validation accuracy: \t', np.average(valid_accuracy))
```
