### Convert to numerical values
Below I'll convert all the variables to numerical values useable in the models

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
var_mod = ['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Type_Combined','Outlet_Type', 
           'Outlet_Identifier', 'Item_Price_Range']
le = LabelEncoder()
for i in var_mod:
    data[i] = le.fit_transform(data[i])
```

```python
#One Hot Coding:
data = pd.get_dummies(data, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Type_Combined','Outlet_Type', 
           'Outlet_Identifier', 'Item_Price_Range'])
```

### Fit models
I can now fit some models and see what works best. I'm going to leave the test data, as I don't have the target value 'Item_Outlet_Sales' for this, so don't know how well each model has done.

I will create a generic 'modelfit' function, which I can then pass in the model I want to test out on, and it will output the accuracy of the model.

```python
from sklearn import cross_validation, metrics
def modelfit(alg, train_X, train_y):
    #Fit the algorithm on the data
    alg.fit(train_X, train_y)
        
    #Predict using the new model on the training set:
    predictions = alg.predict(train_X)

    #Perform cross-validation to see accuracy of model:
    cv_score = cross_validation.cross_val_score(alg, train_X, train_y, cv=20, scoring='mean_squared_error')
    cv_score = np.sqrt(np.abs(cv_score))
    
    #Print model report:
    print ("\nModel Report")
    print ("RMSE : %.4g" % np.sqrt(metrics.mean_squared_error(train_y.values, predictions)))
    print ("CV Score : Mean - %.4g | Std - %.4g | Min - %.4g | Max - %.4g" % (np.mean(cv_score),np.std(cv_score),
                                                                              np.min(cv_score),np.max(cv_score)))
```

### Linear Regression
I'll fit the model, then plot the co-efficients (corresponding to the weighting) of each variable:

```python
from sklearn.linear_model import LinearRegression
alg = LinearRegression(normalize=True)
modelfit(alg, train_X, train_y)
weighting = pd.Series(data=alg.coef_[0], index=train_X.columns).sort_values()
plt.figure(figsize=(12,6))
weighting.plot(kind='bar', title='Model Coefficients')
```

You can see the Outlet Type and Outlet Identifiers have a big effect on the Item_Outlet_Sales, whereas Item_Weight and Item_Visibility have little effect.

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
alg = DecisionTreeRegressor(max_depth=40, min_samples_leaf=150)
modelfit(alg, train_X, train_y)
coef3 = pd.Series(alg.feature_importances_, train_X.columns).sort_values(ascending=False)
plt.figure(figsize=(12,6))
coef3.plot(kind='bar', title='Feature Importances')
```

To optimise this I could cycle over different depth and min leaf sizes.

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
alg = RandomForestRegressor(n_estimators=200,max_depth=10, min_samples_leaf=150,n_jobs=4)
modelfit(alg, train_X, train_y)
coef3 = pd.Series(alg.feature_importances_, train_X.columns).sort_values(ascending=False)
plt.figure(figsize=(12,6))
coef3.plot(kind='bar', title='Feature Importances')
```
