```python
#importing libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the data
data=pd.read_csv('../data/data_cleaned.csv')
```

```python
#shape of the data
data.shape
```

```python
#first five rows of the data
data.head()
```

```python
#checking missing values in the data
data.isnull().sum()
```

```python
#seperating independent and dependent variables
y = data['Survived']
X = data.drop(['Survived'], axis=1)
```

```python
#importing train_test_split to create validation set
from sklearn.model_selection import train_test_split
```

```python
#creating the train and validation set
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 101, stratify=y, test_size=0.25)
```

```python
# distribution in training set
y_train.value_counts(normalize=True)
```

```python
# distribution in validation set
y_valid.value_counts(normalize=True)
```

```python
#shape of training set
X_train.shape, y_train.shape
```

```python
#shape of validation set
X_valid.shape, y_valid.shape
```

```python
#importing decision tree classifier 
from sklearn.tree import DecisionTreeClassifier
```

```python
# how to import decision tree regressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#creating the decision tree function
dt_model = DecisionTreeClassifier(random_state=10)
```

```python
#fitting the model
dt_model.fit(X_train, y_train)
```

```python
#checking the training score
dt_model.score(X_train, y_train)
```

```python
#checking the validation score
dt_model.score(X_valid, y_valid)
```

```python
#predictions on validation set
dt_model.predict(X_valid)
```

```python
dt_model.predict_proba(X_valid) # 0.5 is the default threshhold
```

```python
y_pred = dt_model.predict_proba(X_valid)[:,1] # Get the probabilities of being a one (the 2nd col above)
y_pred[1:10]
```

```python
y_new = []
for i in range(len(y_pred)):
    if y_pred[i]<=0.7:
        y_new.append(0)
    else:
        y_new.append(1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_valid, y_new) # does not dramtically help accuracy.  Shows how to adjust threshold
```

## Changing the max_depth

```python
train_accuracy = []
validation_accuracy = []
for depth in range(1,10):
    dt_model = DecisionTreeClassifier(max_depth=depth, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_valid, y_valid))
```

```python
frame = pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker='o', label = 'Training')
plt.plot(frame['max_depth'], frame['valid_acc'], marker='o', label = 'Validation')
plt.xlabel('Depth of tree')
plt.ylabel('Accuracy')
plt.title('Train v Validation Accuracy\nComparing Decision Tree Depth')
plt.legend()
```

* max_leaf_nodes
* min_samples_split

```python
# Depth = 8 provided highest accuracy
dt_model = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=25, random_state=10)
```

```python
#fitting the model
dt_model.fit(X_train, y_train)
```

```python
#Training score
dt_model.score(X_train, y_train)
```

```python
#Validation score
dt_model.score(X_valid, y_valid)
```
